# 1. Import Libraries

```python
import pandas as pd
import numpy as np

# Display more columns
pd.set_option('display.max_columns', None)
```

# 2. Load datasets

```python
deaths = pd.read_csv("CovidDeaths.csv")
vaccinations = pd.read_csv("CovidVaccinations.csv")
```

```python
deaths.head()
```

```python
df = deaths[[
    "location",
    "date",
    "total_cases",
    "new_cases",
    "total_deaths",
    "population"
]]

df = df.sort_values(["location","date"])

df.head()
```

```python
df = deaths[deaths["continent"].notna()]

df = df.sort_values(["location","date"])

df.head()
```

### Likelihood of Dying if Infected

```python
df = deaths[deaths["location"] == "Algeria"].copy()

df["death_percentage"] = (
    df["total_deaths"] / df["total_cases"]
) * 100

df = df.sort_values(["location","date"])

df[["location","date","total_cases","total_deaths","death_percentage"]].head()
```

### Percentage of Population Infected

```python
df["covid_percentage"] = (
    deaths["total_cases"] /
    deaths["population"]
) * 100

df = df[
    ["location","date","population","total_cases","covid_percentage"]
].sort_values(["location","date"])

df.head()
```

### Countries with Highest Infection Rate

```python
df = (
    deaths.groupby(["location","population"], as_index=False)
    .agg(
        HighestInfectionCount=("total_cases","max")
    )
)

df["percent_of_population_infected"] = (
    df["HighestInfectionCount"] /
    df["population"]
) * 100

df.sort_values(
    "percent_of_population_infected",
    ascending=False
).head(10)
```

### Countries with Highest Death Count

```python
df["total_deaths"] = pd.to_numeric(
    deaths["total_deaths"], errors="coerce"
)

df = (
    deaths.groupby("location", as_index=False)
    .agg(total_death_count=("total_deaths","max"))
)

df.sort_values("total_death_count", ascending=False).head(10)
```

### Death Count by Continent

```python
df = deaths[
    deaths["continent"].notna()
]

df = (
    df.groupby("continent", as_index=False)
    .agg(total_death_count=("total_deaths","max"))
)

df.sort_values(
    "total_death_count",
    ascending=False
)
```

### Global Numbers

```python
deaths["new_deaths"] = pd.to_numeric(
    deaths["new_deaths"], errors="coerce"
)

global_data = deaths[deaths["continent"].notna()]

total_cases = global_data["new_cases"].sum()
total_deaths = global_data["new_deaths"].sum()

death_percentage = (total_deaths / total_cases) * 100

print("Total Cases:", total_cases)
print("Total Deaths:", total_deaths)
print("Death Percentage:", death_percentage)
```

### Vaccination Table

```python
vaccinations.head()
```

### Population vs Vaccinations (JOIN)

```python
merged = pd.merge(
    deaths,
    vaccinations,
    on=["location","date"],
    how="inner"
)

merged.head()
```

### Rolling People Vaccinated (Window Function)

```python
merged["new_vaccinations"] = pd.to_numeric(
    vaccinations["new_vaccinations"], errors="coerce"
)

merged = merged.sort_values(["location","date"])

merged["RollingPeopleVaccinated"] = (
    merged.groupby("location")["new_vaccinations"]
    .cumsum()
)
```

### Percentage of Population Vaccinated (CTE Equivalent)

```python
merged["percent_population_vaccinated"] = (
    merged["RollingPeopleVaccinated"] /
    merged["population"]
) * 100

merged[
    ["location","date","population",
     "RollingPeopleVaccinated",
     "percent_population_vaccinated"]
].head()
```

### Temporary Table Equivalent

```python
percent_population_vaccinated = pd.merge(
    deaths[["continent","location","date","population"]],
    merged[["location","date","new_vaccinations","RollingPeopleVaccinated"]],
    on=["location","date"]
)

percent_population_vaccinated.head()
```

### View Equivalent

```python
PercentPopulationVaccinated = percent_population_vaccinated.copy()

PercentPopulationVaccinated.head()
```

## SQL vs Pandas Concepts

| SQL             | Pandas                 |
| --------------- | ---------------------- |
| SELECT          | Column selection       |
| WHERE           | Boolean filtering      |
| ORDER BY        | `sort_values()`        |
| GROUP BY        | `groupby()`            |
| JOIN            | `merge()`              |
| SUM()           | `.sum()`               |
| MAX()           | `.max()`               |
| Window Function | `groupby().cumsum()`   |
| CTE             | Intermediate DataFrame |
| Temp Table      | New DataFrame          |
| View            | Saved DataFrame        |
